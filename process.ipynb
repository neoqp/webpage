```python
import pandas as pd
df = pd.read_csv('trash_data.csv')

dummy = []
for i in range(len(df)):
    dummy.append({'type' : df.iloc[i]['type'],
                  'amount(g)' : int(df.iloc[i]['amount'].split('g/')[0]),
                  'amount(%)' : int(df.iloc[i]['amount'].split('g/')[1][:-1]),
                  'date' : df.iloc[i]['date']})
    
    dummy.append({'type' : df.iloc[i]['type.1'],
                  'amount(g)' : int(df.iloc[i]['amount.1'].split('g/')[0]),
                  'amount(%)' : int(df.iloc[i]['amount.1'].split('g/')[1][:-1]),
                  'date' : df.iloc[i]['date.1']})
    
    dummy.append({'type' : df.iloc[i]['type.2'],
                  'amount(g)' : int(df.iloc[i]['amount.2'].split('g/')[0]),
                  'amount(%)' : int(df.iloc[i]['amount.2'].split('g/')[1][:-1]),
                  'date' : df.iloc[i]['date.2']})
    
    dummy.append({'type' : df.iloc[i]['type.3'],
                  'amount(g)' : int(df.iloc[i]['amount.3'].split('g/')[0]),
                  'amount(%)' : int(df.iloc[i]['amount.3'].split('g/')[1][:-1]),
                  'date' : df.iloc[i]['date.3']})
    
    dummy.append({'type' : 'CO2',
                  'amount(g)' : int(df.iloc[i]['CO2(gC02)']),
                  'amount(%)' : 0,
                  'date' : df.iloc[i]['date.4']})
    
for i in range(10):
    print(dummy[i])
```
